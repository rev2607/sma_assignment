```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('/content/social_media_data.csv')




data['Engagement Rate (%)'] = (data['Engagements'] / data['Reach']) * 100
monthly_engagement_rate = data['Engagement Rate (%)'].mean()

start_followers = data['Followers'].iloc[0]
end_followers = data['Followers'].iloc[-1]
follower_growth = ((end_followers - start_followers) / start_followers) * 100


top_posts = data.nlargest(3, 'Engagement Rate (%)')


demographics = {
    'Age Groups': ['18-24', '25-34', '35-44', '45-54', '55+'],
    'Percentage': [30, 40, 15, 10, 5]
}
demographics_df = pd.DataFrame(demographics)


plt.figure(figsize=(8, 6))
plt.bar(demographics_df['Age Groups'], demographics_df['Percentage'], color='skyblue')
plt.title('Demographic Breakdown of Followers')
plt.xlabel('Age Groups')
plt.ylabel('Percentage (%)')
plt.show()


print("=== Social Media Metrics Analysis ===")
print(f"1. Average Engagement Rate: {monthly_engagement_rate:.2f}%")
print(f"2. Follower Growth: {follower_growth:.2f}%")
print("\n3. Top-Performing Posts:")
print(top_posts[['Date', 'Post ID', 'Content Type', 'Engagement Rate (%)']])
print("\n4. Demographics Breakdown:")
print(demographics_df)


print("\n=== Recommendations ===")
print("1. Focus on posting more video content as they tend to have higher engagement.")
print("2. Target the 18-34 age group with tailored campaigns as they form 70% of the audience.")
print("3. Experiment with interactive content such as polls and Q&A to improve engagement.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/social_media_content.csv')

data['Engagements'] = data['Likes'] + data['Comments'] + data['Shares']
data['Engagement Rate (%)'] = (data['Engagements'] / data['Reach']) * 100

content_performance = data.groupby('Content Type').agg({
    'Engagements': 'sum',
    'Engagement Rate (%)': 'mean',
    'Post ID': 'count'
}).reset_index()

content_performance.rename(columns={'Post ID': 'Post Count'}, inplace=True)

best_content = content_performance.sort_values(by='Engagement Rate (%)', ascending=False).head(1)

plt.figure(figsize=(8, 6))
plt.bar(content_performance['Content Type'], content_performance['Engagement Rate (%)'], color='skyblue')
plt.title('Average Engagement Rate by Content Type')
plt.xlabel('Content Type')
plt.ylabel('Engagement Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("=== Content Strategy Evaluation ===")
print("\nContent Performance Summary:")
print(content_performance)

print("\nBest-Performing Content Type:")
print(best_content)

print("\n=== Recommendations ===")
if best_content['Content Type'].iloc[0] == 'Promotional':
    print("Focus on creating more promotional content, as it generates high engagement.")
elif best_content['Content Type'].iloc[0] == 'Educational':
    print("Expand the educational content strategy to address audience needs.")
elif best_content['Content Type'].iloc[0] == 'Interactive':
    print("Increase interactive content such as polls, quizzes, and Q&A to boost engagement.")
else:
    print("Leverage the success of this content type while experimenting with other formats.")
```
